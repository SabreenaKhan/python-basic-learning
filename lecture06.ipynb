```python
#Create two sets and find their union. 

setA={1,2,3,4}
setB={5,6,7,8}
print(setA.union(setB))
```

```python
#Create two sets and find their intersection. 
setA={2,4,5,6,7,8}
setB={1,2,3,8,9}
print(setA.intersection(setB))
```

```python
# Create two sets and find their difference (A âˆ’ B). 
print(setA.difference(setB))
```

```python
#Use .copy() to create a copy of a set and modify the copy. 
setA={1,2,3,4,5}
set_copy=setA.copy()
set_copy.pop()
print(setA)
print(set_copy)
```

```python
# Use .clear() to empty a set. 
set_copy.clear()
print(set_copy)
```

```python
# Convert a list with duplicates into a set to remove duplicates. 

li=[1,2,3,4,5,5,6,7,3,2,1]
conv_set=set(li)
print(conv_set)
```

```python
#Write a program to find all unique characters in a string using a set. 
str=input("Enter any string")
st=set(str)
print(st)
```

```python
#Given two lists of students (cricket and football), find students who play both sports (intersection). 
cricket=["alice","bob","john"]
football=["bob","john","sam"]
ckt_set=set(cricket)
ftb_set=set(football)
print(ckt_set.intersection(ftb_set))
```

```python
#Given two lists of students, find students who play only cricket (difference). 

print(ckt_set.difference(ftb_set))
```

```python
sentence=input("enter a sentence")
words_list=sentence.split()
word_set=set(words_list)
print(word_set)
```

```python
#Write a program to check if two sets are equal (ignoring order). 
setA={1,2,3,4,5}
setB={2,3,4,5,6}
print(setA==setB)
```

```python
# Create two sets and check if they have at least one element in common (hint: use intersection). 
common=setA.intersection(setB)
if(len(common)>0):
    print("Common element found")
else:
    print("common element not found")
```
