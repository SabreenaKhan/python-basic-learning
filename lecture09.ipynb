READ AND WRITE MODE(r++)

```python
#Open the file in r+ mode and update the first studentâ€™s name from "Amit" to "Amit Kumar". 
with open(r"C:\Users\Sabreena Khan\Desktop\student.txt.txt","r+")as file:
    content=file.read()
    file.seek(0)
    file.write(content.replace("Sneha","Amit Kumar"))
    file.truncate()
```

```python
#Use r+ mode to move the cursor to the end of the file and add a new student "Sneha, 20, Hyderabad"
with open(r"C:\Users\Sabreena Khan\Desktop\student.txt.txt","r+")as file:
    content=file.read()
    file.write("sneha,20,hyderabad")
```
