DICTIONARIES

```python
#Create a dictionary of 5 countries and their capitals. Print it. 

world_capitals={"India":"new delhi",
                "Japan":"tokyo",
                "france":"paris",
                "brazil":"brasilia",
                "canada":"ottawa"}
print(world_capitals)
```

```python
#Access the capital of "India" from the dictionary. 

print(world_capitals["India"])
```

```python
#Add a new key-value pair "Japan": "Tokyo" to the dictionary. 
world_capitals["USA"]="washington"
print(world_capitals)
```

```python
#Update the capital of "USA" from "Washington" to "New York". 
world_capitals["USA"]="new york"
print(world_capitals)
```

```python
#Delete the key "France" using del. 
del world_capitals["france"]
print(world_capitals)
```

```python
#Create a dictionary of students and marks. Use .keys() to print all student names. 
st_data={
    "john":25,
    "alice":56,
    "bob":98,
    "sam":67
}
students=st_data.keys()
print(students)
```

```python
#Use .values() to print all marks. 
marks=st_data.values()
print(marks)
```

```python
#Use .items() to print all key-value pairs. 
items=st_data.items()
print(items)
```

```python
#Use .get() to access the value of "Rahul" safely (if not found, return "Not Found"). 
print(st_data.get("rahul","not found"))
```

```python
#Use .update() to add multiple key-value pairs in one step. 
user_profile={
    "username":"coder_99",
    "level":5
}
user_profile.update({
    "language":"python",
    "is_active":"true",
    "streak":12
})
print(user_profile)
```

```python
#Use .pop(key) to remove "Germany" from the dictionary. 
countries={
    "india":"new delhi",
    "germany":"berlin",
    "japan":"tokyo"
}
print(countries.pop("germany","not found"))
print(countries)
```

```python
# Try popping a non-existing key using .pop() with a default value. 
print(countries.pop("turkey","not found"))
print(countries)
```

```python
#Use .clear() to remove all elements from the dictionary. 
countries.clear()
print(countries)
```

```python
# Create a dictionary, make a copy using .copy(), and update the copy without affecting the original. 
student_info={
    "john":56,
    "alice":78
}
copy=student_info.copy()
copy.update({
    "sam":56
})
print(student_info)
print(copy)
```

```python
#Count the frequency of words in a sentence using a dictionary
sentence="my name is sabreena khan"
words=sentence.lower().split()
frequency={}
for word in words:
    if word in frequency:
        frequency[word]+=1
    else:
        frequency[word]=1
print(frequency)
```

```python
# Store product names as keys and prices as values. Print products costing more than 100. 
products={
    "chair":3000,
    "table":350,
    "chips":20,
    "chocolate":10
}
for product in products:
    if(products[product]>100):
        print(product)
```

```python
# Write a program to check if "Apple" exists as a key in a dictionary. 
fruits={
    "apple":350,
    "mango":500,
    "banana":250
}
if "apple" in fruits:
    print("yes apple in fruits")
else:
    print("no")
          
```

```python
#Find the key with the maximum value in a dictionary of student marks
student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "Diana": 95,
    "Ethan": 88
}
top_student=max(student_marks,key=student_marks.get)
print(top_student)
```

```python
# Convert two lists into a dictionary (e.g., ["name","age","city"] and ["Amit",25,"Delhi"]). 
keys=["name","age","city"]
values=["amit",25,"delhi"]
result_dict=dict(zip(keys,values))
print(result_dict)
```

```python
#Swap keys and values in a dictionary.
og_dict={
    "A":1,
    "B":2,
    "C":3
}
swapped_dict={value:key for key,value in og_dict.items()}
print(swapped_dict)
```

```python
#Create a dictionary with two keys: "names" and "ages", storing 3 students. 

std={
    "names":["amit","neha","rahul"],
    "age":[20,22,19]
}
```

```python
#Print the first student’s name and age. 

print(f"name = {std["names"][0]}")
print(f"age= {std["age"][0]}")
```

```python
#Add a new student "Priya", 21 to both lists. 
std["names"].append("Priya")
std["age"].append(21)
print(std)
```

```python
#Update "Rahul"’s age to 20.
std["age"][2]=20
print(std)
```

```python
#Remove "Neha" and her age from both lists.
idx=std["names"].index("neha")
std["names"].pop(idx)
std["age"].pop(idx)
print(std)
```

```python
#Find the student with the maximum age
max_age=max(std["age"])
max_age_idx=std["age"].index(max_age)
max_student=std["names"][max_age_idx]
print(f" {max_student} is the student with maximum age of {max_age} ")
```

```python
#Print all students older than 20.
for i in range(len(std["age"])):
    if std["age"][i]>20:
        print(f" name = {std["names"][i]} and age= {std["age"][i]}")
```

```python
#Convert the dictionary of lists into a list of dictionaries.
students_list = [{"name": n, "age": a} for n, a in zip(std["names"], std["age"])]
print(students_list)
```

```python
#Convert the dictionary of lists into a Pandas DataFrame 
import pandas as pd
df=pd.DataFrame(students_list)
print(df)
```

```python
#Sort students by age using the dictionary of lists
print(df.sort_values(by="age"))
```
