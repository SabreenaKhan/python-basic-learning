```python
#Write a function that takes a list of numbers and returns their average.
def fun5(num_list):
    total_sum=sum(num_list)
    count=len(num_list)
    return total_sum/count
my_list=[1,2,3,4]
fun5(my_list)
```

```python
#Write a function that counts vowels in a string.
def fun6(str):
    count=0
    for char in str:
        if char in 'aeiouAEIOU':
            count+=1
    print(count)
fun6("sabreena")
```

```python
#Write a function that calculates factorial of a number (using loop)
def fact(num):
    factorial=1
    while(num>0):
        factorial*=num
        num-=1
    print(factorial)
fact(4)
```

```python
# Write a function to check if a string is a palindrome. 
def palindrome(str):
    str=str.lower()
    left=0
    right=len(str)-1
    while left<right:
        if str[left]!=str[right]:
            return False
        left+=1
        right-=1
    return True
palindrome("saas")
```

RECURSIVE FUNCTIONS

```python
#Write a recursive function to calculate factorial. 
def factorial(num):
    if num==0:
        return 1
    else:
        return (num *(factorial(num-1)))
    
print(factorial(4))
```

```python
# Write a recursive function to print Fibonacci series up to n terms.
def fibo(n):
    if n<=0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
n=10
for i in range(n):
    print(fibo(i),end=" ")
    
```

```python
#Write a function that takes a list of numbers and returns the largest element.
def largest(num_list):
    return max(num_list)
numbers=[2,3,4,5,6]
print(largest(numbers))
```

```python
#Write a function that takes a dictionary and prints all key-value pairs.
def dicti(student):
    for key,value in student.items():
        print(f"{key}:{value}")

students={
    "alice":20,
    "max":19
}
dicti(students)
```

```python
#Write a function that takes a set and returns its length. 
def get_length(my_set):
    return len(my_set)
my_set={"apple","mango","orange"}
print(get_length(my_set))
```

```python
#Write a function that takes a list of dictionaries (students with marks) and returns the topperâ€™s name.
def list_of_dic(li):
    topper=li[0]
    for student in li:
        if student["marks"]>topper["marks"]:
            topper=student
    return topper["name"]

student_list=[
    {"name":"amit","marks":20},
    {"name":"max","marks":21},
    {"name":"el","marks":23}
]
print(f"The topper is {list_of_dic(student_list)}")
```
