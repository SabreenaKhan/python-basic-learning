LOOPS

```python
#Write a program to print numbers from 1 to 10 using a for loop. 
for i in range(1,11):
    print(i)
```

```python
#Find the sum of first 10 natural numbers using a while loop. 
i=1
sum=0
while(i<=10):
    sum+=i
    i+=1
print(sum)
```

```python
#Take a number as input and print its multiplication table up to 10. 
num=int(input("enter a number"))
for i in range(1,11):
    print(f"{num} X {i} = {num*i}")
```

```python
#Find the factorial of a number using a while loop. 
num=int(input("Enter a number"))
i=1
fact=1;
while(i<=num):
    fact*=i
    i+=1
print(fact)
```

```python
#reverse a number
num=int(input("enter a number"))
rev=0
while(num>0):
    digit=num%10
    rev=(rev*10) +digit
    num=num//10

print(rev)
```

```python
for i in range(1,50):
    if(i%2==0):
        print(i)
```

```python
#sum of digits
num=int(input("enter a number"))
sum=0
while(num>0):
    digit=num%10
    sum+=digit
    num//=10
print(sum)
```

```python
#fibonacci series
a=0
b=1
for i in range (10):
    print(a , end=" ")
    temp=a
    a=b
    b=temp+b
```
